```python
import pandas as pd

url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)

df.head()
```

```python
df["medv"]
```

```python
df["medv"].max()
```

```python
df["medv"].min()
```

```python
df["medv"].mean()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["rm"], df["medv"])
plt.xlabel("Número de habitaciones")
plt.ylabel("Precio medio de la vivienda (medv)")
plt.title("Relación entre habitaciones y precio de vivienda en Boston")
plt.show()
```

```python
df.groupby("rm")["medv"].mean()
```

```python
df.groupby("crim")["medv"].mean()
```

```python
plt.scatter(df["crim"], df["medv"])
plt.xlabel("Criminalidad")
plt.ylabel("Precio")
plt.title("Relación entre la tasa de criminalidad y el precio medio de vivienda en Boston")
plt.show
```

```python
plt.scatter(df["rm"], df["medv"], c=df["crim"], cmap="viridis")
plt.xlabel("Número de habitaciones")
plt.ylabel("Precio medio de la vivienda (medv)")
plt.title("Habitaciones vs Precio con criminalidad codificada por color")
plt.colorbar(label="Tasa de criminalidad")
plt.show()
```
